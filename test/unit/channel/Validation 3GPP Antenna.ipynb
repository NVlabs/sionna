# Radiation pattern of an antenna element

```python
from sionna.channel.tr38901 import AntennaElement

antenna_element = AntennaElement("38.901", 0, dtype=tf.complex64)
antenna_element.show()

antenna_element = AntennaElement("omni", 0, dtype=tf.complex64)
antenna_element.show()
```

# Geometry of a panel

```python
from sionna.channel.tr38901 import AntennaPanel

antenna_panel = AntennaPanel(num_rows=8,
                             num_cols=4,
                             polarization='dual',
                             vertical_spacing=0.5,
                             horizontal_spacing=0.5,
                             dtype=tf.complex64)
antenna_panel.show()
```

# Geometry of a panel array

```python
from sionna.channel.tr38901 import PanelArray

array = PanelArray(num_rows_per_panel=4,
                   num_cols_per_panel=4,
                   polarization='dual',
                   polarization_type='VH',
                   antenna_pattern='38.901',
                   carrier_frequency=3.5e9,
                   num_cols=4,
                   num_rows=4,
                   panel_horizontal_spacing=3,
                   panel_vertical_spacing=3,
                   dtype=tf.complex64)
array.show()
```
